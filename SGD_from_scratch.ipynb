# **Algorithm**

Pick an initial random point x0.

x1 = x0 - r [(df/dx) of x0]

x2 = x1- r [(df/dx) of x1]

Similarly, we find for x0, x1, x2 …………. x[k-1]

Where r is our learning rate and 
df/dx the gradient function to minimize our loss
